# Context
A telecom company wants to use their historical customer data to predict behaviour to retain customers. You can analyse all relevant customer data and develop focused customer retention programs.

# Data Description
Each row represents a customer, each column contains customer’s attributes described on the column Metadata. The data set includes information about:

- Customers who left within the last month – the column is called Churn
- Services that each customer has signed up for – phone, multiple lines, internet, online security, online backup, device protection, tech support, and streaming TV and movies
- Customer account information – how long they’ve been a customer, contract, payment method, paperless billing, monthly
charges, and total charges
- Demographic info about customers – gender, age range, and if they have partners and dependents

# Project Objective
Build a model that will help to identify the potential customers who have a higher probability to churn. This help the company to understand the pinpoints and patterns of customer churn and will increase the focus on strategising customer retention.

## Import and Warehouse data

```python
# Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
import pickle

# To enable plotting graphs in Jupyter notebook
%matplotlib inline
```

```python
# Read the csv file
churnData = pd.read_csv("TelcomCustomer-Churn.csv")
churnData.head()
```

```python
# Explore the shape of the data frame
churnData.shape
```

## Data Cleansing

```python
# Check the datatypes
churnData.info()
```

- SeniorCitizen is of type int and should be converted to categorical
- TotalCharges is of type object and should be be converted to float
- Other columns are of object type and should be converted to categorical

```python
# TotalCharges has a few blank values. Replace blank values to zero and convert to float
churnData['TotalCharges'] = churnData.TotalCharges.replace(' ','0')
churnData['TotalCharges'] = churnData.TotalCharges.astype(float)
```

```python
# CustomerID is not required, so drop the column
churnData.drop(['customerID'], axis= 1, inplace=True)
```

```python
# Convert columns of object type to categorical
for feature in churnData.columns: # Loop through all columns in the dataframe
    if churnData[feature].dtype == 'object': # Check if the datatype is of object data type
        churnData[feature] = pd.Categorical(churnData[feature]) # Convert to categorical
```

```python
# Check if the data contains missing or NaN values
churnData.isna().sum()
```

 - None of the columns have null values

```python
# Function to print unique value counts
def print_value_counts(s):
    print(churnData[s].value_counts())
    print('-----------------------------------------------')

print_value_counts("gender")
print_value_counts("SeniorCitizen")
print_value_counts("Partner")
print_value_counts("Dependents")
print_value_counts("PhoneService")
print_value_counts("MultipleLines")
print_value_counts("InternetService")
print_value_counts("OnlineSecurity")
print_value_counts("OnlineBackup")
print_value_counts("DeviceProtection")
print_value_counts("TechSupport")
print_value_counts("StreamingTV")
print_value_counts("StreamingMovies")
print_value_counts("Contract")
print_value_counts("PaperlessBilling")
print_value_counts("PaymentMethod")
```

- One hot encode the following categorical columns
1. gender
2. SeniorCitizen
3. Partner
4. Dependents
5. PhoneService
6. MultipleLines
7. InternetService
8. OnlineSecurity
9. OnlineBackup
10. DeviceProtection
11. TechSupport
12. StreamingTV
13. StreamingMovies
14. Contract
15. PaperlessBilling
16. PaymentMethod

- Replace Churn values with '0' for 'No' and '1' for 'Yes

```python
# Churn values to be replaced
replaceStruct = {
                    "Churn" : {"No": 0, "Yes":1}
                }

# List of columns for one hot encoding
oneHotCols = ["gender","SeniorCitizen","Partner","Dependents","PhoneService","MultipleLines","InternetService","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","StreamingMovies","Contract","PaperlessBilling","PaymentMethod"]
```

```python
# Make a copy of the data frame
churnDataCopy = churnData

# Replace string categorical values to integer values
churnData = churnData.replace(replaceStruct)

# One hot encoding
churnData = pd.get_dummies(churnData,columns = oneHotCols)

churnData.info()
```

## Data pre-processing

```python
# Create dataframe of predictor columns and target class
X = churnData.drop('Churn',axis = 1) # Predictors
y = churnData['Churn'] # Target Class - 0 or 1
```

```python
# Split the dataset into training and test datasets in the ratio 70:30
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
# Check split of target class

print("Customers churned in the original dataset           : {0} ({1:0.2f}%)".format(len(churnData.loc[churnData['Churn'] == 1.0]), (len(churnData.loc[churnData['Churn'] == 1])/len(churnData.index)) * 100))
print("Customers who did not churn in the original dataset : {0} ({1:0.2f}%)".format(len(churnData.loc[churnData['Churn'] == 0.0]), (len(churnData.loc[churnData['Churn'] == 0])/len(churnData.index)) * 100))
print("")

print("Customers churned in the training dataset           : {0} ({1:0.2f}%)".format(len(y_train[y_train[:] == 1]), (len(y_train[y_train[:] == 1])/len(y_train)) * 100))
print("Customers who did not churn in the training dataset : {0} ({1:0.2f}%)".format(len(y_train[y_train[:] == 0]), (len(y_train[y_train[:] == 0])/len(y_train)) * 100))
print("")

print("Customers churned in the test dataset               : {0} ({1:0.2f}%)".format(len(y_test[y_test[:] == 1]), (len(y_test[y_test[:] == 1])/len(y_test)) * 100))
print("Customers who did not churn in the test dataset     : {0} ({1:0.2f}%)".format(len(y_test[y_test[:] == 0]), (len(y_test[y_test[:] == 0])/len(y_test)) * 100))
```

- The split of customers churned and not churned is approximately 26% and 74% respectively. The model will be more effective for predictions for customers who don't churn

## Model building, testing and tuning

```python
# Plot heatmap of confusion matrix
def plot_cm(test, predict):
    cm=metrics.confusion_matrix(test, predict, labels=[1, 0])
    df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],columns = [i for i in ["Predict 1","Predict 0"]])
    plt.figure(figsize = (7,5))
    sns.heatmap(df_cm, annot=True,fmt=".0f",cmap="Spectral")

# List to maintain model scores
list_scores = []
```

### Decision Tree

```python
# Create decision tree classifier
dTree = DecisionTreeClassifier()
```

```python
# Fit the model on training data
dTree.fit(X_train,y_train)
```

```python
# Use the model to predict using test data
y_predict = dTree.predict(X_test)
```

```python
# Get the model score on training data
score_train = dTree.score(X_train, y_train)
print('Score for Decison Tree on the training data is %1.4f' %score_train)

# Get the model score on test data
score_test = dTree.score(X_test, y_test)
print('Score for Decision Tree on the test data is %1.4f' %score_test)

list_scores.append(['Decision Tree',score_test])
```

```python
# Plot heatmap of confusion matrix
plot_cm(y_test, y_predict)
```

```python
# Print the classification report
print(metrics.classification_report(y_test,y_predict))
```

### Decision Tree with hyper parameter tuning

```python
# Create decision tree classifier
dTreeR = DecisionTreeClassifier(criterion = 'gini', max_depth = 3, random_state=1)

# Fit the model on training data
dTreeR.fit(X_train, y_train)
```

```python
# Use the model to predict using test data
y_predict = dTreeR.predict(X_test)
```

```python
# Get the model score on training data
score_train = dTreeR.score(X_train, y_train)
print('Score for Decision Tree(regularized) on the training data is %1.4f' %score_train)

# Get the model score on test data
score_test = dTreeR.score(X_test, y_test)
print('Score for Decision Tree(regularized) on the test data is %1.4f' %score_test)

list_scores.append(['Decision Tree - Regularized',score_test])
```

```python
# Plot heatmap of confusion matrix
plot_cm(y_test,y_predict)
```

```python
# Print the classification report
print(metrics.classification_report(y_test,y_predict))
```
