```python
# GUI for predicting customer churn
 
# import everything from tkinter module 
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk

import numpy as np
import pandas as pd
import pickle

# Function to predict churn based on user inputs through GUI
def predict():
    #Global dataframe to capture the inputs for the model
    global mod_X 
    mod_X = pd.DataFrame()

    # Populate inputs from user to the dataframe
    get_input(mod_X)    
    
    # Unpickle the model
    loaded_model = pickle.load(open('AdaBoostClassifier.sav', 'rb'))
    
    # Predict using the unpickled model
    result = loaded_model.predict(mod_X)
        
    if(result == 0): 
        churn.set('No') # Model prediction is 0, set churn to 'No'
    else:
        churn.set('Yes') # Model prediction is 1, set churn to 'Yes'
        
# Function to preprocess data captured from GUI and populate the dataframe
def get_input(mod_X):    

    # Set Tenure
    try:
        tenure = int(tenureInMonths.get())        
    except:                
        tenureInMonths.set(0)
        messagebox.showerror("showerror", "Enter a valid value for Tenure")
        raise
    
    # Set Monthly Charges
    try:
        mCharges = float(monthlyCharges.get())                
    except:
        monthlyCharges.set(0)
        messagebox.showerror("showerror", "Enter a valid value for Monthly Charges")
        raise
    
    # Set Total Charges
    try:
        tCharges = float(totalCharges.get())
    except:
        totalCharges.set(0)
        messagebox.showerror("showerror", "Enter a valid value for Total Charges")
        raise 

    # Set Gender
    if(gender.get() == "Female"):
        gender_Female = 1
        gender_Male = 0
    else:
        gender_Female = 0
        gender_Male = 1
    
    # Set Senior Citizen
    if(seniorCitizen.get() == "0"):
        seniorCitizen_0 = 1
        seniorCitizen_1 = 0
    else:
        seniorCitizen_0 = 0
        seniorCitizen_1 = 1
    
    # Set Partner
    if(partner.get() == "No"):
        partner_No = 1
        partner_Yes = 0
    else:
        partner_No = 0
        partner_Yes = 1
    
    # Set Dependents
    if(dependents.get() == "No"):
        dependents_No = 1
        dependents_Yes = 0
    else:
        dependents_No = 0
        dependents_Yes = 1
        
    # Set Phone Service    
    if(phoneService.get() == "No"):
        phService_No = 1
        phService_Yes = 0
    else:
        phService_No = 0
        phService_Yes = 1

    # Set Multiple Lines
    if(multipleLines.get() == "No"):
        multLines_No = 1
        multLines_Yes = 0
        multLines_No_phService = 0
    elif(multipleLines.get() == "Yes"):
        multLines_No = 0
        multLines_Yes = 1
        multLines_No_phService = 0
    else:
        multLines_No = 0
        multLines_Yes = 0
        multLines_No_phService = 1
    
    # Set Internet Service
    if(internetService.get() == "No"):
        intService_No = 1
        intService_DSL = 0
        intService_fibreOptic = 0
    elif(internetService.get() == "DSL"):
        intService_No = 0
        intService_DSL = 1
        intService_fibreOptic = 0
    else: 
        intService_No = 0
        intService_DSL = 0
        intService_fibreOptic = 1

    # Set Online Security
    if(onlineSecurity.get() == "No"):
        onSecurity_No = 1
        onSecurity_Yes = 0
        onSecurity_No_intService = 0
    elif(onlineSecurity.get() == "Yes"):
        onSecurity_No = 0
        onSecurity_Yes = 1
        onSecurity_No_intService = 0
    else:
        onSecurity_No = 0
        onSecurity_Yes = 0
        onSecurity_No_intService = 1
    
    # Set Online Backup
    if(onlineBackup.get() == "No"):
        onBackup_No = 1
        onBackup_Yes = 0
        onBackup_No_intService = 0
    elif(onlineBackup.get() == "Yes"):
        onBackup_No = 0
        onBackup_Yes = 1
        onBackup_No_intService = 0
    else:
        onBackup_No = 0
        onBackup_Yes = 0
        onBackup_No_intService = 1
    
    # Set Device Protection
    if(deviceProtection.get() == "No"):
        deviceProtect_No = 1
        deviceProtect_Yes = 0
        deviceProtect_No_intService = 0
    elif(deviceProtection.get() == "Yes"):
        deviceProtect_No = 0
        deviceProtect_Yes = 1
        deviceProtect_No_intService = 0
    else:
        deviceProtect_No = 0
        deviceProtect_Yes = 0
        deviceProtect_No_intService = 1

    # Set Tech Support
    if(techSupport.get() == "No"):
        techSup_No = 1
        techSup_Yes = 0
        techSup_No_intService = 0
    elif(techSupport.get() == "Yes"):
        techSup_No = 0
        techSup_Yes = 1
        techSup_No_intService = 0
    else:
        techSup_No = 0
        techSup_Yes = 0
        techSup_No_intService = 1
    
    # Set Streaming TV
    if(streamingTV.get() == "No"):
        sTV_No = 1
        sTV_Yes = 0
        sTV_No_intService = 0
    elif(streamingTV.get() == "Yes"):
        sTV_No = 0
        sTV_Yes = 1
        sTV_No_intService = 0
    else:
        sTV_No = 0
        sTV_Yes = 0
        sTV_No_intService = 1

    # Set Streaming Movies
    if(streamingMovies.get() == "No"):
        sMovies_No = 1
        sMovies_Yes = 0
        sMovies_No_intService = 0
    elif(streamingMovies.get() == "Yes"):
        sMovies_No = 0
        sMovies_Yes = 1
        sMovies_No_intService = 0
    else:
        sMovies_No = 1
        sMovies_Yes = 0
        sMovies_No_intService = 0

        
    # Set Contract
    if(contract.get() == "Month-to-month"):
        cont_Monthly = 1
        cont_oneYear = 0
        cont_twoYear = 0
    elif(streamingMovies.get() == "One year"):
        cont_Monthly = 0
        cont_oneYear = 1
        cont_twoYear = 0
    else:
        cont_Monthly = 0
        cont_oneYear = 0
        cont_twoYear = 1
    
    # Set Paperless Billing
    if(paperlessBilling.get() == "No"):
        paperlessBilling_No = 1     
        paperlessBilling_Yes = 0
    else:  
        paperlessBilling_No = 0     
        paperlessBilling_Yes = 1
    
    # Set Payment Method
    if(paymentMethod.get() == "Bank transfer (automatic)"):
        paymentBT = 1
        paymentCC = 0
        paymentEC = 0
        paymentMC = 0
    elif(paymentMethod.get() == "Credit card (automatic)"):
        paymentBT = 0
        paymentCC = 1
        paymentEC = 0
        paymentMC = 0
    elif(paymentMethod.get() == "Electronic check"):
        paymentBT = 0
        paymentCC = 0
        paymentEC = 1
        paymentMC = 0
    elif(paymentMethod.get() == "Mailed check"):
        paymentBT = 0
        paymentCC = 0
        paymentEC = 0
        paymentMC = 1
    
    # Populate the dataframe with pre-processed data
    mod_X.loc[0,'tenure']= tenure
    mod_X.loc[0,'MonthlyCharges']= mCharges
    mod_X.loc[0,'TotalCharges']= tCharges 
    mod_X.loc[0,'gender_Female']= gender_Female
    mod_X.loc[0,'gender_Male']= gender_Male
    mod_X.loc[0,'SeniorCitizen_0']=seniorCitizen_0
    mod_X.loc[0,'SeniorCitizen_1']= seniorCitizen_1
    mod_X.loc[0,'Partner_No']= partner_No
    mod_X.loc[0,'Partner_Yes']= partner_Yes
    mod_X.loc[0,'Dependents_No']= dependents_No
    mod_X.loc[0,'Dependents_Yes']= dependents_Yes
    mod_X.loc[0,'PhoneService_No']= phService_No
    mod_X.loc[0,'PhoneService_Yes']= phService_Yes
    mod_X.loc[0,'MultipleLines_No']= multLines_No
    mod_X.loc[0,'MultipleLines_No phone service']= multLines_No_phService
    mod_X.loc[0,'MultipleLines_Yes']= multLines_Yes
    mod_X.loc[0,'InternetService_DSL']= intService_DSL
    mod_X.loc[0,'InternetService_Fiber optic']= intService_fibreOptic
    mod_X.loc[0,'InternetService_No']= intService_No
    mod_X.loc[0,'OnlineSecurity_No']= onSecurity_No
    mod_X.loc[0,'OnlineSecurity_No internet service']= onSecurity_No_intService
    mod_X.loc[0,'OnlineSecurity_Yes']= onSecurity_Yes
    mod_X.loc[0,'OnlineBackup_No']= onBackup_No
    mod_X.loc[0,'OnlineBackup_No internet service']= onBackup_No_intService
    mod_X.loc[0,'OnlineBackup_Yes']= onBackup_Yes
    mod_X.loc[0,'DeviceProtection_No']= deviceProtect_No
    mod_X.loc[0,'DeviceProtection_No internet service']= deviceProtect_No_intService
    mod_X.loc[0,'DeviceProtection_Yes']= deviceProtect_Yes
    mod_X.loc[0,'TechSupport_No']= techSup_No
    mod_X.loc[0,'TechSupport_No internet service']= techSup_No_intService
    mod_X.loc[0,'TechSupport_Yes']= techSup_Yes
    mod_X.loc[0,'StreamingTV_No']= sTV_No
    mod_X.loc[0,'StreamingTV_No internet service']= sTV_No_intService
    mod_X.loc[0,'StreamingTV_Yes']= sTV_Yes
    mod_X.loc[0,'StreamingMovies_No']= sMovies_No
    mod_X.loc[0,'StreamingMovies_No internet service']= sMovies_No_intService
    mod_X.loc[0,'StreamingMovies_Yes']= sMovies_Yes
    mod_X.loc[0,'Contract_Month-to-month']= cont_Monthly
    mod_X.loc[0,'Contract_One year']= cont_oneYear
    mod_X.loc[0,'Contract_Two year']= cont_twoYear
    mod_X.loc[0,'PaperlessBilling_No']= paperlessBilling_No
    mod_X.loc[0,'PaperlessBilling_Yes']= paperlessBilling_Yes
    mod_X.loc[0,'PaymentMethod_Bank transfer (automatic)']= paymentBT
    mod_X.loc[0,'PaymentMethod_Credit card (automatic)']= paymentCC
    mod_X.loc[0,'PaymentMethod_Electronic check']= paymentEC
    mod_X.loc[0,'PaymentMethod_Mailed check']= paymentMC

# Set default values for option menu and text entry components
def setDefault():
    gender.set("Female")
    seniorCitizen.set("0")
    partner.set("No")
    dependents.set("No")
    tenureInMonths.set(0)
    phoneService.set("No")
    multipleLines.set("No")
    internetService.set("No")
    onlineSecurity.set("No")
    onlineBackup.set("No")
    deviceProtection.set("No")
    techSupport.set("No")
    streamingTV.set("No")
    streamingMovies.set("No")
    contract.set("Month-to-month")
    paperlessBilling.set("No")
    paymentMethod.set("Bank transfer (automatic)")
    monthlyCharges.set(0)
    totalCharges.set(0)
    churn.set('')


# Create a GUI window 
win = tk.Tk()

win.geometry('300x500')
 
# set the title of GUI window 
win.title("Telecom Customer Churn - Prediction") 
    
# Gender
label1 = ttk.Label(win,text="Gender")
label1.grid(row=1, column=1,sticky=tk.W)

gender = tk.StringVar()
optGender = ttk.OptionMenu(win,gender,"Female","Male")
optGender.grid(row=1,column=2,sticky=tk.W)

# Senior Citizen
label2 = ttk.Label(win,text="Senior Citizen")
label2.grid(row=2, column=1,sticky=tk.W)

seniorCitizen = tk.StringVar()
optSeniorCitizen = ttk.OptionMenu(win,seniorCitizen,"0","1")
optSeniorCitizen.grid(row=2,column=2,sticky=tk.W)

# Partner
label3 = ttk.Label(win,text="Partner")
label3.grid(row=3, column=1,sticky=tk.W)

partner = tk.StringVar()
optPartner = ttk.OptionMenu(win,partner,"No","Yes")
optPartner.grid(row=3,column=2,sticky=tk.W)

# Dependents
label4 = ttk.Label(win,text="Dependents")
label4.grid(row=4, column=1,sticky=tk.W)

dependents = tk.StringVar()
optDependents = ttk.OptionMenu(win,dependents,"No","Yes")
optDependents.grid(row=4,column=2,sticky=tk.W)

# Tenure
label5 = ttk.Label(win,text="Tenure in months")
label5.grid(row=5, column=1,sticky=tk.W)

tenureInMonths = tk.IntVar()
entryTenure = ttk.Entry(win,textvariable = tenureInMonths)
entryTenure.grid(row=5,column=2,sticky=tk.W)

# Phone Service
label6 = ttk.Label(win,text="Phone Service")
label6.grid(row=6, column=1,sticky=tk.W)

phoneService = tk.StringVar()
optPhoneService = ttk.OptionMenu(win,phoneService,"No","Yes")
optPhoneService.grid(row=6,column=2,sticky=tk.W)

# Multiple Lines
label7 = ttk.Label(win,text="Multiple Lines")
label7.grid(row=7, column=1,sticky=tk.W)

multipleLines = tk.StringVar()
optMultipleLines = ttk.OptionMenu(win,multipleLines,"No","Yes","No phone service")
optMultipleLines.grid(row=7,column=2,sticky=tk.W)

# Internet Service
label8 = ttk.Label(win,text="Internet Service")
label8.grid(row=8, column=1,sticky=tk.W)

internetService = tk.StringVar()
optInternetService = ttk.OptionMenu(win,internetService,"No","DSL","Fiber optic")
optInternetService.grid(row=8,column=2,sticky=tk.W)

# Online Security
label9 = ttk.Label(win,text="Online Security")
label9.grid(row=9, column=1,sticky=tk.W)

onlineSecurity = tk.StringVar()
optOnlineSecurity = ttk.OptionMenu(win,onlineSecurity,"No","Yes","No internet service")
optOnlineSecurity.grid(row=9,column=2,sticky=tk.W)
                                  
# Online Backup
label10 = ttk.Label(win,text="Online Backup")
label10.grid(row=10, column=1,sticky=tk.W)

onlineBackup = tk.StringVar()
optOnlineBackup = ttk.OptionMenu(win,onlineBackup,"No","Yes","No internet service")
optOnlineBackup.grid(row=10,column=2,sticky=tk.W)    

# Device Protection
label11 = ttk.Label(win,text="Device Protection")
label11.grid(row=11, column=1,sticky=tk.W)

deviceProtection = tk.StringVar()
optDeviceProtection = ttk.OptionMenu(win,deviceProtection,"No","Yes","No internet service")
optDeviceProtection.grid(row=11,column=2,sticky=tk.W)    

# Tech Support
label12 = ttk.Label(win,text="Tech Support")
label12.grid(row=12, column=1,sticky=tk.W)

techSupport = tk.StringVar()
optTechSupport = ttk.OptionMenu(win,techSupport,"No","Yes","No internet service")
optTechSupport.grid(row=12,column=2,sticky=tk.W)    

# Streaming TV
label13 = ttk.Label(win,text="Streaming TV")
label13.grid(row=13, column=1,sticky=tk.W)

streamingTV = tk.StringVar()
optStreamingTV = ttk.OptionMenu(win,streamingTV,"No","Yes","No internet service")
optStreamingTV.grid(row=13,column=2,sticky=tk.W)    

# Streaming Movies
label14 = ttk.Label(win,text="Streaming Movies")
label14.grid(row=14, column=1,sticky=tk.W)

streamingMovies = tk.StringVar()
optStreamingMovies = ttk.OptionMenu(win,streamingMovies,"No","Yes","No internet service")
optStreamingMovies.grid(row=14,column=2,sticky=tk.W)    

# Contract
label15 = ttk.Label(win,text="Contract")
label15.grid(row=15, column=1,sticky=tk.W)

contract = tk.StringVar()
optContract = ttk.OptionMenu(win,contract,"Month-to-month","One year","Two year")
optContract.grid(row=15,column=2,sticky=tk.W)    

# Paperless Billing
label16 = ttk.Label(win,text="Paperless Billing")
label16.grid(row=16, column=1,sticky=tk.W)

paperlessBilling = tk.StringVar()
optPaperlessBilling = ttk.OptionMenu(win,paperlessBilling,"No","Yes")
optPaperlessBilling.grid(row=16,column=2,sticky=tk.W)    

# Payment Method
label17 = ttk.Label(win,text="Payment Method")
label17.grid(row=17, column=1,sticky=tk.W)

paymentMethod = tk.StringVar()
optPaymentMethod = ttk.OptionMenu(win,paymentMethod,"Bank transfer (automatic)","Credit card (automatic)","Electronic check","Mailed check")
optPaymentMethod.grid(row=17,column=2,sticky=tk.W)   

# Monthly Charges
label18 = ttk.Label(win,text="Monthly Charges")
label18.grid(row=18, column=1,sticky=tk.W)

monthlyCharges = tk.DoubleVar()
entryMonthlyCharges = ttk.Entry(win,textvariable = monthlyCharges)
entryMonthlyCharges.grid(row=18,column=2,sticky=tk.W)

# Total Charges
label19 = ttk.Label(win,text="Total Charges")
label19.grid(row=19, column=1,sticky=tk.W)

totalCharges = tk.DoubleVar()
entryTotalCharges = ttk.Entry(win,textvariable = totalCharges)
entryTotalCharges.grid(row=19,column=2,sticky=tk.W)

# Churn Prediction
label20 = ttk.Label(win,text="Churn Prediction")
label20.grid(row=20, column=1,sticky=tk.W)

global churn
churn = tk.StringVar()
churnPrediction = ttk.Entry(win,textvariable = churn)
churnPrediction.grid(row=20,column=2,sticky=tk.W)


button1 = ttk.Button(win, text ="Clear",command=setDefault)
button1.grid(row=24, column = 1)

button2 = ttk.Button(win, text ="Predict",command=predict)
button2.grid(row=24, column = 2)

setDefault()

# start the GUI 
win.mainloop() 
```
