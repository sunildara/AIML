### BUSINESS CONTEXT<br>
Recognising multi-digit numbers in photographs captured at street level is an important component of modern-day map making. A classic example of a corpus of such street-level photographs is Googleâ€™s Street View imagery composed of hundreds of millions of geo-located 360-degree panoramic images. The ability to automatically transcribe an address number from a geo-located patch of pixels and associate the transcribed number with a known street address helps pinpoint, with a high degree of accuracy, the location of the building it represents. More broadly, recognising numbers in photographs is a problem of interest to the optical character recognition community. While OCR on constrained domains like document processing is well studied, arbitrary multi-character text recognition in photographs is still highly challenging. This difficulty arises due to the wide variability in the visual appearance of text in the wild on account of a large range of fonts, colours, styles, orientations, and character arrangements. The recognition problem is further complicated by environmental factors such as lighting, shadows, specularity, and occlusions as well as by image acquisition factors such as resolution, motion, and focus blurs. In this project, we will use the dataset with images centred around a single digit (many of the images do contain some distractors at the sides). Although we are taking a sample of the data which is simpler, it is more complex than MNIST because of the distractors.
<br><br>

### Data pre-processing<br>
Reshape training,validation and test sets into compatible shapes<br>
Sequential model in tensorflow.keras expects data to be in the format (n_e, n_h, n_w, n_c)<br>
n_e= number of examples, n_h = height, n_w = width, n_c = number of channels<br>
do not reshape labels<br>

```python
# Reshape training,validation and test sets into compatible shapes
X_train = X_train.reshape(X_train.shape[0],32,32,1)
X_val = X_val.reshape(X_val.shape[0],32,32,1)
X_test =  X_test.reshape(X_test.shape[0],32,32,1)
```

### Normalize data<br>
Normalize our data by dividing the RGB codes with 255 (which is the maximum RGB code minus the minimum RGB code)<br>
Make sure that the values are float so that we can get decimal points after division<br>

```python
X_train = X_train.astype('float32')
X_val = X_val.astype('float32')
X_test = X_test.astype('float32')

X_train /= 255
X_val /= 255
X_test /= 255
```

```python
# Print Max and Min values in training dataset
print("Max value in X_train:", X_train.max())
print("Min value in X_train:", X_train.min(),'\n')
#print('\n')
# Print Max and Min values in validation dataset
print("Max value in X_train:", X_val.max())
print("Min value in X_train:", X_val.min(),'\n')

# Print Max and Min values in test dataset
print("Max value in X_train:", X_test.max())
print("Min value in X_train:", X_test.min())
```

```python
from sklearn.decomposition import PCA
x_train_pca = PCA(n_components=1024)
x_train_pca_result = x_train_pca.fit_transform(X_train_df[feat_cols].values)
```

```python
x_train_pca_result.shape
```

Building nural network with 3 layers of each 512 nurons and relu as activation function.<br>
10 nurons as last layer for output, with softmax activation function.
Considered SGD as optimizer with loss function as perameter.

```python
def modelNN(le_rate,x_train_set,y_train_set,loss_fun):
    opz = tf.keras.optimizers.SGD( learning_rate=le_rate, nesterov=True)
    model = tf.keras.Sequential(name="my_sequential")
    model.add(layers.Dense(1024, activation="relu", name="layer1",use_bias= True))
    model.add(BatchNormalization())
    model.add(layers.Dense(1024, activation="relu", name="layer2",use_bias= True))
    model.add(layers.Dense(1024, activation="relu", name="layer3",use_bias= True))
    model.add(BatchNormalization())
    model.add(layers.Dense(1024, activation="relu", name="layer4",use_bias= True))
    model.add(layers.Dense(10, name="layer5",activation="softmax",use_bias= True))
    model.compile(optimizer=opz, loss=loss_fun, metrics=['accuracy'])
    model.fit(x_train_set, y_train_set, epochs=10, batch_size=1000)
    return model;
```

Calling above method with different loss function related to classification, with this will get to know which loss function at which learning rate provide better accurecy and low loss.

```python
train_metrics =[]
loss_funs = ['categorical_crossentropy','poisson']
for lf in loss_funs:
    for lr in arange(0.02,0.07,0.01):
        final_model = modelNN(lr,x_train_pca_result,y_train,lf)
        train_op = final_model.evaluate(x_train_pca_result, y_train, verbose=0)
        train_metrics.append([lf,lr,train_op[0],train_op[1]])
```

```python
df_train_result = pd.DataFrame(train_metrics,columns=['loss_fun','learning_rate','loss','accuracy'])
df_train_result["learning_rate"] = df_train_result["learning_rate"].astype(float)
df_train_result["loss"] = df_train_result["loss"].astype(float)
df_train_result["accuracy"] = df_train_result["accuracy"].astype(float)
sns.lineplot(x=df_train_result["learning_rate"],y=df_train_result["accuracy"],hue=df_train_result["loss_fun"])
```

#### Prepare Validation data applying PCA

```python
X_val_img = []
for i in range(0,60000):
    img = X_val[i].flatten()
    X_val_img.append(img)
X_val_img = np.array(X_val_img)
X_val_df = pd.DataFrame(X_val_img,columns=feat_cols)

# Applying PCA with only 512 columns to increase processing efficiency 
x_val_pca = PCA(n_components=1024)
x_val_pca_result = x_val_pca.fit_transform(X_val_df[feat_cols].values)
```

Based on above observation in accuracy and loss according to the learning rate, we will build the final model with following parameters.<br><br>
Total 4 layers<br>
3 intermediate layers of each 512 nurons and relu as activation function.<br>
10 nurons as last layer for output, with softmax activation function.<br> 
Considered SGD as optimizer<br>
Loss as categorical_crossentropy<br>
learning rate 0.5 <br>
50 epochs

```python
    opz = tf.keras.optimizers.SGD( learning_rate=0.06, nesterov=True)
    final_model = tf.keras.Sequential(name="final_seq_model")
    final_model.add(layers.Dense(1024, activation="relu", name="layer1",use_bias= True))
    final_model.add(BatchNormalization())
    final_model.add(layers.Dense(1024, activation="relu", name="layer2",use_bias= True))
    final_model.add(BatchNormalization())
    final_model.add(layers.Dense(512, activation="relu", name="layer3",use_bias= True))
    final_model.add(BatchNormalization())
    final_model.add(layers.Dense(1024, activation="relu", name="layer4",use_bias= True))
    final_model.add(layers.Dense(10, name="layer5",activation="softmax",use_bias= True))
    final_model.compile(optimizer=opz, loss="categorical_crossentropy", metrics=['accuracy'])
```

```python
# Final model fit
history = final_model.fit(x=x_train_pca_result,y=y_train, epochs=30, batch_size=1000,validation_data=(x_val_pca_result,y_val))
```

```python
# Prepare test data
X_tst_img = []
for i in range(0,18000):
    img = X_test[i].flatten()
    X_tst_img.append(img)
X_tst_img = np.array(X_tst_img)
X_tst_df = pd.DataFrame(X_tst_img,columns=feat_cols)

# Applying PCA with only 512 columns to increase processing efficiency 
x_tst_pca = PCA(n_components=1024)
x_tst_pca_result = x_tst_pca.fit_transform(X_tst_df[feat_cols].values)
```
