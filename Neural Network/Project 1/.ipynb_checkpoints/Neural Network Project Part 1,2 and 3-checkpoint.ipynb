## Design, train, tune and test a neural network regressor. 

```python
sigdata.head(5)
```

```python
x_data = sigdata.drop(["Signal_Strength"],axis=1)
```

```python
x_data.head(5)
```

```python
y_data = sigdata["Signal_Strength"]
```

```python
x_data_z = x_data.apply(zscore)

from sklearn.decomposition import PCA
pca = PCA(n_components = x_data_z.shape[1])
pca.fit(x_data_z)
```

```python
# Bar plot to plot the explained variance ratio by individual principal components
plt.bar(list(range(1, x_data_z.shape[1]+1)),pca.explained_variance_ratio_,alpha=0.5, align='center')
plt.ylabel('Variation explained')
plt.xlabel('Principal Component')
plt.show()
```

```python
# Step plot to plot the cumulative explained variance ratio
plt.step(list(range(1, 13)),np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('Cum of variation explained')
plt.xlabel('Principal Component')
plt.show()
```

From the above plot, it is clear that more than 95% of the variance is explained by 9 components. Let's transform the features with 9 principal components

```python
x_train,x_test,y_train,y_test = train_test_split(x_data_z,y_data,test_size=0.3,random_state=1)
```

```python
# PCA with 9 components
pca9 = PCA(n_components=9, random_state=2)

# Fit the resampled training data
pca9.fit(x_train)

# Transform the resampled training data
X_train_pca = pca9.transform(x_train)

# Transform the test data
X_test_pca = pca9.transform(x_test)
```

```python
y_train=tf.keras.utils.to_categorical(
    y_train, num_classes=9, dtype='float32'
)
```

```python
y_test=tf.keras.utils.to_categorical(
    y_test, num_classes=None, dtype='float32'
)
```

```python
def modelNN(le_rate,x_train_set,y_train_set,loss_fun):
    opz = tf.keras.optimizers.SGD( learning_rate=le_rate, nesterov=True)
    model = tf.keras.Sequential(name="my_sequential")
    model.add(layers.Flatten())
    model.add(layers.Dense(18, activation="relu", name="layer1",use_bias= True))
    model.add(layers.Dense(18, activation="relu", name="layer2",use_bias= True))
    model.add(layers.Dense(9, name="layer3",activation="softmax",use_bias= True))
    model.compile(optimizer=opz, loss=loss_fun, metrics=['accuracy'])
    model.fit(X_train_pca, y_train, epochs=10, batch_size=10)
    return model;
```

```python
train_metrics =[]
test_metrics=[]
loss_funs = ['mean_squared_error','MeanAbsoluteError','MeanAbsolutePercentageError','MeanSquaredLogarithmicError','CosineSimilarity']
for lf in loss_funs:
    for lr in arange(0.01,0.09,0.01):
        final_model = modelNN(lr,X_train_pca,y_train,lf)
        train_op = final_model.evaluate(X_train_pca, y_train, verbose=0)
        test_op = final_model.evaluate(X_test_pca, y_test, verbose=0)
        train_metrics.append([lf,lr,train_op[0],train_op[1]])
        test_metrics.append([lf,lr,test_op[0],test_op[1]])
```

```python
df_train_result = pd.DataFrame(train_metrics,columns=['loss_fun','learning_rate','loss','accuracy'])
```

```python
df_train_result["learning_rate"] = df_train_result["learning_rate"].astype(float)
df_train_result["loss"] = df_train_result["loss"].astype(float)
df_train_result["accuracy"] = df_train_result["accuracy"].astype(float)
```

```python
sns.lineplot(x=df_train_result["learning_rate"],y=df_train_result["accuracy"],hue=df_train_result["loss_fun"],)
```

```python
sns.lineplot(x=df_train_result["learning_rate"],y=df_train_result["loss"],hue=df_train_result["loss_fun"])
```

Based on above comparisions between loss functions CosineSimilarity is providing high accuracy and low loss, will train the final model with CosineSimilarity for regression at 0.045 learning rate. Considering 0.045 because there is no large varience in accurecy or loss after this value.

```python
model_final = modelNN(0.045,X_train_pca,y_train,'CosineSimilarity')
```

```python
model_final.evaluate(X_test_pca, y_test, verbose=0)
```

Loss is negligible and accuracy is 0.597 on test data.

### Save the model

```python
#FInal model save to file
model_final.save("reg_model.h5")
```

```python
# load model
model_from_file = load_model('reg_model.h5')
```

```python
#evaluate the model from file.
model_from_file.evaluate(X_test_pca, y_test, verbose=0)
```

# Part two of the project

***• DOMAIN:*** Electronics and Telecommunication <br>
***• CONTEXT:*** A communications equipment manufacturing company has a product which is responsible for emitting
informative signals. Company wants to build a machine learning model which can help the company to predict the
equipment’s signal quality using various parameters.<br>
***• DATA DESCRIPTION:*** The data set contains information on various signal tests performed:<br>
    1. Parameters: Various measurable signal parameters.<br>
    2. Signal_Quality: Final signal strength or quality<br>
***• PROJECT OBJECTIVE:*** The need is to build a classifier which can use these parameters to determine the signal strength or 
quality [as number].

```python
#Import data
sigdata_clf = pd.read_csv("Part- 1,2&3 - Signal.csv")
```

```python
#Data analysis & visualisation
#Analysis is done above for the same data.
#splitting the data and appling the zscore
x_data_c = sigdata_clf.drop(["Signal_Strength"],axis=1)
y_data_c = sigdata_clf["Signal_Strength"]
x_train_c,x_test_c,y_train_c,y_test_c = train_test_split(x_data_c,y_data_c,test_size=0.3,random_state=1)
x_train_cz = x_train_c.apply(zscore)
x_test_cz = x_test_c.apply(zscore)
```

```python
sigdata_clf['Signal_Strength'].value_counts()
```

```python
y_train_cc=tf.keras.utils.to_categorical(
    y_train_c, num_classes=9, dtype='int'
)
```

```python
y_test_cc=tf.keras.utils.to_categorical(
    y_test_c, num_classes=9, dtype='int'
)
```

```python
y_train_cc.shape
```

```python
x_train_c.shape
```

```python
def clf_model(opt,X,Y,LF):
    model = tf.keras.Sequential(name="clasification_NN")
    model.add(layers.Flatten())
    model.add(layers.Dense(20, activation="relu", name="layer2",use_bias= True))
    model.add(layers.Dense(20, activation="relu", name="layer3",use_bias= True))
    model.add(layers.Dense(9, name="layer4",activation="softmax",use_bias= True))
    model.compile(optimizer=opt, loss=LF, metrics=['accuracy'])
    model.fit(X, Y, epochs=10, batch_size=10)
    return model;
```

```python
#Considering only 3 types of losses as only these 3 loss functions are good for categorical classification and learning rate from 0.01 to 0.09 increment of 0.01
clf_train_met =[]
loss_funs_lst = ['categorical_crossentropy','poisson','kl_divergence']
for lf in loss_funs_lst:
    for lr in arange(0.01,0.09,0.01):
        opt = tf.keras.optimizers.SGD( learning_rate=lr, nesterov=True)
        clf_model1 = clf_model(opt,x_train_cz,y_train_cc,lf)
        train_op = clf_model1.evaluate(x_train_cz, y_train_cc, verbose=0)
        clf_train_met.append([lf,lr,train_op[0],train_op[1]])
```

```python
df_train_clf = pd.DataFrame(clf_train_met,columns=['loss_fun','learning_rate','loss','accuracy'])
df_train_clf["learning_rate"] = df_train_clf["learning_rate"].astype(float)
df_train_clf["loss"] = df_train_clf["loss"].astype(float)
df_train_clf["accuracy"] = df_train_clf["accuracy"].astype(float)
fig, axarr = plt.subplots(2, 2, figsize=(12, 8))
sns.lineplot(x=df_train_clf["learning_rate"],y=df_train_clf["accuracy"],hue=df_train_clf["loss_fun"],ax=axarr[0][0])
sns.lineplot(x=df_train_clf["learning_rate"],y=df_train_clf["loss"],hue=df_train_clf["loss_fun"],ax=axarr[0][1])
plt.suptitle('Accuracy VS learning Rate in SGD')
```

Comparing in between these 3 loss functions with SGD optimizer, poisson loss function is giving consistent accurecy with less loss, we will try with diffenet optimizers also to check the performence.

```python
#Considering only 3 types of losses as only these 3 loss functions are good for categorical classification
#and learning rate from 0.01 to 0.09 increment of 0.01 using Adam optimizer
clf_train_adam =[]
loss_funs_adam = ['categorical_crossentropy','poisson','kl_divergence']
for lf in loss_funs_adam:
    for lr in arange(0.01,0.09,0.01):
        opt = tf.keras.optimizers.Adam( learning_rate=lr)
        clf_model1 = clf_model(opt,x_train_cz,y_train_cc,lf)
        train_op = clf_model1.evaluate(x_train_cz, y_train_cc, verbose=0)
        clf_train_adam.append([lf,lr,train_op[0],train_op[1]])
```

```python
df_train_adam = pd.DataFrame(clf_train_adam,columns=['loss_fun','learning_rate','loss','accuracy'])
df_train_adam["learning_rate"] = df_train_adam["learning_rate"].astype(float)
df_train_adam["loss"] = df_train_adam["loss"].astype(float)
df_train_adam["accuracy"] = df_train_adam["accuracy"].astype(float)
fig, axarr = plt.subplots(2, 2, figsize=(12, 8))
sns.lineplot(x=df_train_adam["learning_rate"],y=df_train_adam["accuracy"],hue=df_train_adam["loss_fun"],ax=axarr[0][0])
sns.lineplot(x=df_train_adam["learning_rate"],y=df_train_adam["loss"],hue=df_train_adam["loss_fun"],ax=axarr[0][1])
plt.suptitle('Accuracy VS learning Rate in Adam')
```

There is no much difference between SGD and Adam optimizers, accurecy and loss are same. Will be using SGD for final model building with poisson as loss function and learning rate of 0.08

```python
opt = tf.keras.optimizers.SGD( 0.08, nesterov=True)
clf_final_model = clf_model(opt,x_train_cz,y_train_cc,'poisson')
```

Accurecy on train data # 58%

```python
#Accurecy on test data
clf_final_model.evaluate(x_test_cz,y_test_cc)
```

### Save the final model
