**Distribution of 'bmi', 'age' and 'charges'**

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'under-weight[(bmi<18.5)]'
    elif bmi >= 18.5 and bmi <= 24.9:
        return 'normal-weight[bmi[18.5-24.9]]'
    elif bmi > 24.9 and bmi <= 30.0:
        return 'over-weight[bmi(24.9-30.0])]'
    elif bmi > 30.0:
        return "obese[(bmi >30)]"
mData['cbmi'] = mData['bmi'].apply(lambda x: "none")
for idx, row in mData.iterrows():
        mData.at[idx, 'cbmi'] = bmi_category(row['bmi'])
        by_bmi = mData.groupby(by='cbmi').size()
print (by_bmi)
print ('-'*100)
```

```python
def age_category(age):
    age_dict = {
        0: '0-9',
        1: '10-19',
        2: '20-29',
        3: '30-39',
        4: '40-49',
        5: '50-59',
        6: '60-69',
        7: '70-79',
        8: '80-89',
        9: '90-99',
        10: '100-200'
    }
    return age_dict[age//10]
    
mData['cage'] = mData['age'].apply(lambda x: "none")
for idx, row in mData.iterrows():
    mData.at[idx, 'cage'] = age_category(row['age'])
    by_age = mData.groupby(by='cage').size()
print (by_age)
print ('-'*100)
```

**Inferential Analysis by Statistical Significance Tests**

****T-test****

**To test whether the charges of people who smoke differ significantly from the people who do not**
* H0 : Mu(smoker charges) = Mu(non-smoker charges)
* H1 : Mu(smoker charges) != Mu(non-smoker charges)


```python
import scipy.stats as stats
```

```python
mData_smoker = np.array(mData[mData.smoker == 'yes'].charges)
mData_non = np.array(mData[mData.smoker == 'no'].charges) 
```

```python
t, pvalue  = stats.ttest_ind(mData_smoker, mData_non, axis = 0)
if pvalue < 0.05: 
    print(f'Reject H0 as the p-value ({(pvalue)}) < 0.05')
else:
    print(f' fail to reject H0 as the p-value ({round(pvalue,6)}) > 0.05')
print('The test confirms that the charges of people who smoke differ significantly from the charges of people\
who do not.')
```

***T-test***

**To test whether the bmi of males differ significantly from that of females**

* H0 : Mu(male bmi) = Mu(female bmi)
* H1 : Mu(male bmi) != Mu(female bmi)

```python
mData_male_bmi   = np.array(mData[mData.sex ==   'male'].bmi)
mData_female_bmi = np.array(mData[mData.sex == 'female'].bmi)
```

```python
t, pvalue = stats.ttest_ind(mData_male_bmi, mData_female_bmi, axis = 0)
if pvalue <0.05:
    print(f'Reject H0 as the p-value ({round(pvalue,5)})< 0.05.')
else:
    print(f'Fail to Reject H0 as the p-value ({round(pvalue,5)} > 0.05.')
print('The test fails to confirm that the male bmi differ significantly from that of female bmi.')
```

**Test of proportions**

* 'sex' and 'smoker' are two categorical variables
* We want to see if the proportion of smokers is significantly different in different genders in the sample 
to infer the population proportion.
#### Ho = The proportion of male & female smokers are equal
#### Ha = The proportion of male & female smokers are not equal

```python
male_smokers   = mData[mData['sex'] == 'male'].smoker.value_counts()[1]   # number of male smokers
female_smokers = mData[mData['sex'] == 'female'].smoker.value_counts()[1] # number of female smokers
n_males   = mData.sex.value_counts()[0]                                   # number of males in the data 
n_females = mData.sex.value_counts()[1]                                   #number of females in the data
```

```python
print([male_smokers, female_smokers],[n_males, n_females])
print(f' Proportion of smokers in males, females = {round(male_smokers/n_males,2)}%, {round(female_smokers/n_females,2)}% \
respectively')
```

The proportions are remarkably different. Let us find whether they are statistically significant?

```python
from statsmodels.stats.proportion import proportions_ztest
stat, pval = proportions_ztest([male_smokers, female_smokers], [n_males, n_females])
if pval < 0.05:
    print(f'     Since pval is {round(pval,4)} less than the significance level of 5% we reject the H0 and conclude that\
 difference in proportion of smokers in males & females is significant.')
else:
    print(f' Since pval is {round(pval,4)} more than the significance level of 5% we fail to reject the H0 and conclude \
that difference in proportion of smokers in males & females is not significant.')
```

**ANOVA Test**

* $H_0$: $\mu0Child$ = $\mu1Child$ = $\mu2Child$
* $H_A$: At least one $\mu$ differs 

```python
mData_Female = mData[~mData['sex'].isin(['male'])]
```

```python
bmi_0C = np.array(mData_Female[mData_Female.children == 0].bmi)
bmi_1C = np.array(mData_Female[mData_Female.children == 1].bmi)
bmi_2C = np.array(mData_Female[mData_Female.children == 2].bmi)

print('Count, Mean, Standard Deviation of bmi - 0 Childern: %3d, %3.2f and %3.2f' %(len(bmi_0C), bmi_0C.mean(), np.std(bmi_0C, ddof =1)))
print('Count, Mean, Standard Deviation of bmi - 1 Childern: %3d, %3.2f and %3.2f' %(len(bmi_1C), bmi_1C.mean(), np.std(bmi_1C, ddof =1)))
print('Count, Mean, Standard Deviation of bmi - 2 Childern: %3d, %3.2f and %3.2f' %(len(bmi_2C), bmi_2C.mean(), np.std(bmi_2C, ddof =1)))
```

```python
consolidated_bmi_df = pd.DataFrame()

df0 = pd.DataFrame({'bmi_type': 'C0', 'Consolidated_bmi': bmi_0C })
df1 = pd.DataFrame({'bmi_type': 'C1', 'Consolidated_bmi': bmi_1C })
df2 = pd.DataFrame({'bmi_type': 'C2', 'Consolidated_bmi': bmi_2C })
consolidated_bmi_df = consolidated_bmi_df.append(df0)
consolidated_bmi_df = consolidated_bmi_df.append(df1)
consolidated_bmi_df = consolidated_bmi_df.append(df2)
```

* $H_0$: $\mu1$ = $\mu2$ = $\mu3$
* $H_A$: At least one $\mu$ differs 

Here $\mu1$, $\mu2$ and $\mu3$ are the mean bmi of women with zero, one and two child/children respectively respectively.

```python
import statsmodels.api as sm
from   statsmodels.formula.api import ols
 
mod = ols('Consolidated_bmi ~ bmi_type', data = consolidated_bmi_df).fit()
aova_table = sm.stats.anova_lm(mod, typ=2)
print(aova_table)
```

p value is 0.715858 and it is more than our chosen level of signifance of 5%. We fail to reject the Null hypothesis.
